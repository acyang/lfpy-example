```python
import plotly.io as pio
import plotly.graph_objects as go
import plotly
import numpy as np
import scipy.io as sio
import LFPy
print(pio.renderers)
#pio.renderers.default = "jupyterlab"
```

```python
X, Y, Z = np.mgrid[-500:501:20, -500:501:20, -400:1201:40]   #51*51*41
print(X.shape, X[2,5,8])
A = X.flatten()
print(A.shape)
B = A.reshape(X.shape)
print(B.shape, B[2,5,8])
```

```python
sio.savemat('test.mat', {'x': X})
lfp_data2 = sio.loadmat('test.mat')
print(lfp_data2.keys())
print(lfp_data2['x'].shape, lfp_data2['x'][2,5,8])
```

```python
lfp_data2 = sio.loadmat('lfp_new.mat')
print(lfp_data2.keys())
print(lfp_data2['x'].flatten().shape, lfp_data2['y'].flatten().shape, lfp_data2['z'].flatten().shape, lfp_data2['lfp'].flatten().shape)
```

```python
cell = LFPy.Cell(morphology = 'morphologies/L5_Mainen96_LFPy.hoc', passive=True)
```

```python
fig = plotly.subplots.make_subplots(rows=1,cols=1,specs=[[{"type": "scene"}]] )
fig.append_trace(go.Volume(
    x=lfp_data2['x'].flatten(),
    y=lfp_data2['y'].flatten(),
    z=lfp_data2['z'].flatten(),
    value= np.log10(lfp_data2['lfp'].flatten()),
    isomin=np.amin(np.log10(lfp_data2['lfp'])),
    isomax=np.amax(np.log10(lfp_data2['lfp'])),
    opacity=0.1, # needs to be small to see through all surfaces
    surface_count=17, # needs to be a large number for good volume rendering
    ),1,1
)

for sec in LFPy.cell.neuron.h.allsec():
    idx = cell.get_idx(sec.name())
    fig.append_trace(go.Scatter3d(
        x=np.r_[cell.xstart[idx], cell.xend[idx][-1]],
        y=np.r_[cell.ystart[idx], cell.yend[idx][-1]],
        z=np.r_[cell.zstart[idx], cell.zend[idx][-1]],
        mode='lines',
        line=dict(
        color='darkblue',
        showscale=False,
        width=1),
        showlegend=False,
        ),1,1
    )
    
fig.show()
```
